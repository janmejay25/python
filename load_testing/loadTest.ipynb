```python
import csv
import matplotlib.pyplot as plt
from collections import Counter
import datetime

timestamps = []
status_codes = []

# Step 1: Load results from CSV
with open("D:/code material/python/results.csv", newline="") as f:
    reader = csv.reader(f)
    next(reader)  # skip header
    for row in reader:
        timestamps.append(float(row[0]))
        status_codes.append(row[1])

# Step 2: Plot request outcomes by time
plt.figure(figsize=(10, 5))
converted_times = [datetime.datetime.fromtimestamp(ts) for ts in timestamps]
colors = ["green" if code == "200" else "red" for code in status_codes]
plt.scatter(converted_times, status_codes, c=colors, label="Status Codes", alpha=0.6)
plt.title("Request Outcomes Over Time")
plt.xlabel("Time")
plt.ylabel("HTTP Status Code")
plt.xticks(rotation=30)
plt.tight_layout()
plt.grid(True)
plt.show()

# Step 3: Status Code Distribution Pie Chart
code_counts = Counter(status_codes)
labels = list(code_counts.keys())
sizes = list(code_counts.values())

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Status Code Distribution")
plt.axis("equal")
plt.show()
```

```python
from collections import defaultdict
import math

rps = defaultdict(int)
for ts in timestamps:
    second = math.floor(ts)
    rps[second] += 1

# Sort by time
sorted_rps = sorted(rps.items())

# Plot RPS over time
plt.figure(figsize=(10, 4))
x, y = zip(*sorted_rps)
x = [datetime.datetime.fromtimestamp(ts) for ts in x]

plt.plot(x, y, color="blue")
plt.title("Requests Per Second Over Time")
plt.xlabel("Time")
plt.ylabel("Requests/sec")
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=30)
plt.show()
```
